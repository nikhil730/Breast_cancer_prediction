```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Breast_cancer_data.csv")
```

```python
df.head()
```

```python
df.drop(df.filter(regex="Unnamed"),axis=1, inplace=True)
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
#Plotting the diagnosis subplots
sns.countplot(x="diagnosis",data=df)
```

```python
#storing the value of the output variable 'Diagnosis' in 'y'
y=df['diagnosis']
y.head()
```

```python
#storing the value of M as 1 and B as 0
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
Y=label_encoder.fit_transform(y)
np.unique(Y)
```

```python
#creating the dataframe for inputs
X=df.drop(labels=["diagnosis","id"],axis=1)
X.head()
```

```python
#scaling the values in the dataset
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X=scaler.transform(X)
X
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
X_train.shape
```

```python
X_test.shape
```

```python
#importing libraries
from keras import Sequential 
from keras.layers import Dense,Activation,Dropout
```

```python
#implemeting ANN
model=Sequential()
model.add(Dense(16,input_dim=30,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1))
model.add(Activation('sigmoid'))

#compiling the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#model summary
print(model.summary())
```

```python
#fitting the model
model.fit(X_train,y_train,verbose=1,epochs=100,batch_size=64,validation_data=(X_test,y_test))
```
